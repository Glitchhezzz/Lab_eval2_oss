```python
import pandas as pd
file_path = (r'C:\Users\22103196\Desktop\Inpatient_Rehabilitation_Facility.csv') 
df = pd.read_csv(file_path)
df['Certification Date'] = pd.to_datetime(df['Certification Date'], format='%Y-%m-%d', errors='coerce')
```

```python
Non_profit_after_2011 = df[(df['Ownership Type'] == 'Non-profit') & (df['Certification Date'] > '2011-10-01')]

print(Non_profit_after_2011)
```

```python
birmingham_profit = df[(df['City/Town'].str.upper() == 'BIRMINGHAM') & (df['Ownership Type'] == 'For profit')]

print(birmingham_profit)
```

```python
df['ZIP Code'] = pd.to_numeric(df['ZIP Code'], errors='coerce') 
zip_range_providers = df[(df['ZIP Code'] >= 85000) & (df['ZIP Code'] <= 90000)]

print(zip_range_providers)
```

```python
ownership_counts = df['Ownership Type'].value_counts()

print(ownership_counts)
```
